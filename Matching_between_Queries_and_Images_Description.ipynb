At the beginning of the matching process, we try to use the glove model which considered the best model in the evaluation test but we get weak scores in the matching so we integrate the pre-trained word2vec model on Google News we try to retrain the pre-trained model Devset then we applied the matching process between the lists of  queries that created by the glove and lists of labels that describe images.

````
#creating a word Embedding 
def get_embedding_vec(word):
    idx = modelG.dictionary.get(word, -1)
    if idx<0:
        #print("Missing word : '%s'" % (word,))
        return np.zeros(  (100, ), dtype='float32')  # UNK
    return modelG.word_vectors[idx]

def get_closest_word(vec, number=5):
    dst = (np.dot(modelG.word_vectors, vec)
                   / np.linalg.norm(modelG.word_vectors, axis=1)
                   / np.linalg.norm(vec))
    word_ids = np.argsort(-dst)
    return [(modelG.inverse_dictionary[x], dst[x]) for x in word_ids[:number]
            if x in modelG.inverse_dictionary]
````
